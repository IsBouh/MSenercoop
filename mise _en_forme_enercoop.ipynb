```python
import pandas as pd
```

```python
data= pd.read_csv("/Users/hanane/Desktop/enercoop/enercorp.csv")
```

```python
data['Catégorie'] = ""
```

```python
## Attribution Labels catégorie
data.loc[data['Unnamed: 0']< 3, "Catégorie"] = "offre"
data.loc[data['Unnamed: 0'] >=3, "Catégorie"] = "souscription"

data.loc[data['Unnamed: 0'] >=12, "Catégorie"] = "sociétaire"

data.loc[data['Unnamed: 0'] >=17, "Catégorie"] = "facturation"

data.loc[data['Unnamed: 0'] ==14, "Catégorie"] = "espace_client"
data.loc[data['Unnamed: 0'] ==15, "Catégorie"] = "espace_client"
data.loc[data['Unnamed: 0'] ==30, "Catégorie"] = "espace_client"

data.loc[data['Unnamed: 0'] ==31, "Catégorie"] = "déménagement"
data.loc[data['Unnamed: 0'] ==32, "Catégorie"] = "déménagement"

data.loc[data['Unnamed: 0'] ==33, "Catégorie"] = "contact"
data.loc[data['Unnamed: 0'] ==11, "Catégorie"] = "contact"

data.loc[data['Unnamed: 0'] ==16, "Catégorie"] = "résiliation"

data.loc[data['Unnamed: 0'] ==34, "Catégorie"] = "intervention"

## Encoding labels catégorie
from sklearn import preprocessing
code= preprocessing.LabelEncoder()
data['Code_Categ']= code.fit_transform(data['Catégorie'])
```

```python
##Mise en forme data
data['Question']= data['question']
data['Reponse']= data['response']
data.drop(['Unnamed: 0','question', 'response'], axis='columns', inplace=True)
```

```python
## Création colonne Client / Pas client
data['Client/Pas_Client'] = "Client"
```

```python
## Attribution labels
data.loc[data['Code_Categ'] == 5 , "Client/Pas_Client"] = "Pas_Client"
data.loc[data['Code_Categ'] == 8, "Client/Pas_Client"] = "Pas_Client"
data.loc[data['Code_Categ'] == 0, "Client/Pas_Client"] = "Indéterminé"

## Encoding Labels
code= preprocessing.LabelEncoder()
data['Code_Customer']= code.fit_transform(data['Client/Pas_Client'])
```
