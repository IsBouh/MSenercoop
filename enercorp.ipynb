```python
import numpy as np
import time
import pandas as pd


# fonction pour générerun temps aléatoire entre les click
def generate_number_delay():
    """
    Retourne un chiffre aléatoire suivant une distribution normal de moyenne 4 secondes et d'ecart-type de 0.8
    Ce chiffre aléatoire est le délai après chaque clic. Le but est de simuler le comportement d'un humain :
    un délai fixe peut attirer l'attention des contrôleurs, tout comme un délai aléatoire d'une distribution uniforme
    """
    mean = 4
    sigma = 0.8
    delai = np.random.normal(mean,sigma,1)[0]
    while delai<0:
        delai = np.random.normal(mean,sigma,1)[0]
    time.sleep(delai)
```

```python
import numpy as np
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import json

dico_Ques_Resp = {}

# fonction pour générerun temps aléatoire entre les click
def generate_number_delay():
    """
    Retourne un chiffre aléatoire suivant une distribution normal de moyenne 4 secondes et d'ecart-type de 0.8
    Ce chiffre aléatoire est le délai après chaque clic. Le but est de simuler le comportement d'un humain :
    un délai fixe peut attirer l'attention des contrôleurs, tout comme un délai aléatoire d'une distribution uniforme
    """
    mean = 2
    sigma = 0.8
    delai = np.random.normal(mean,sigma,1)[0]
    while delai<0:
        delai = np.random.normal(mean,sigma,1)[0]
    time.sleep(delai)


#choix du navigateur 
browser = webdriver.Chrome()

# choix du site 
browser.get('https://www.enercoop.fr/')

#Clic vers le lien FAQ
faq = browser.find_element_by_xpath('//*[@id="block-nrc-main-nrc-main-sticky"]/div/div/div/div/div/div[1]/a')
faq.click()
generate_number_delay()

# 'Je ne suis pas encore client' et 'Je suis déjà client'
question_reponse = browser.find_elements_by_css_selector('li.blocks-item')
question_reponse[0].click()
generate_number_delay()

# section des 
sections = browser.find_elements_by_css_selector("h3.section-tree-title a")

for section in range(len(sections)) :
    sections = browser.find_elements_by_css_selector("h3.section-tree-title a")
    sections[section].click()
    generate_number_delay()
    click_articles = browser.find_elements_by_css_selector('li.article-list-item a')
    for click_article in range(len(click_articles)) :
        click_articles = browser.find_elements_by_css_selector('li.article-list-item a')
        click_articles[click_article].click()
        
        question = browser.find_element_by_tag_name('h1')
        response = browser.find_element_by_class_name('article-body')
        dico_Ques_Resp[question.text] = response.text        
        generate_number_delay()
        browser.back()

    generate_number_delay()
    browser.back()

generate_number_delay()
browser.back()

question_reponse = browser.find_elements_by_css_selector('li.blocks-item')
question_reponse[1].click()

# section des 
sections = browser.find_elements_by_css_selector("h3.section-tree-title a")

for section in range(len(sections)) :
    sections = browser.find_elements_by_css_selector("h3.section-tree-title a")
    sections[section].click()
    generate_number_delay()
    click_articles = browser.find_elements_by_css_selector('li.article-list-item a')
    for click_article in range(len(click_articles)) :
        click_articles = browser.find_elements_by_css_selector('li.article-list-item a')
        click_articles[click_article].click()
        
        question = browser.find_element_by_tag_name('h1')
        response = browser.find_element_by_class_name('article-body')
        dico_Ques_Resp[question.text] = response.text        
        generate_number_delay()
        browser.back()

    generate_number_delay()
    browser.back()

generate_number_delay()
browser.back()


with open('result.json', 'w') as fp:
    json.dump(dico_Ques_Resp, fp)

fin = pd.DataFrame({'question' : [], 'response' : [] })
for key, value in dico_Ques_Resp.items() :
    fin = fin.append({'question' : key, 'response' : value }, ignore_index=True)

fin.to_csv('enercorp.csv')
```

```python
#fonction qui scrape une page : 


def scap_page (liste_raquettes) :
    #je crée ma liste de raquette pour en connaitre la taille
    raquettes = browser.find_elements_by_class_name('thumbnail')
    
    #cette boucle clique sur une raquette, ajoute les infos au df et revient sur la page des raquettes
    for i in range(len(raquettes)) :
        #je recrée ma liste de raquette car la info change quand je change de page
        raquettes = browser.find_elements_by_class_name('thumbnail')
        raquettes[i].click()
    
        # récupération du nom de la raquette
        raquette_name = browser.find_element_by_css_selector('div.productDetails h2')
        myName = raquette_name.text
        
        # récupération des infos détaillés de la raquette 
        tout_key = browser.find_elements_by_tag_name('dt')
        tout_value = browser.find_elements_by_tag_name('dd')
        
        # je crée un dico avec NA en valeur défaut
        dico1 = {'Flex:' : "NA",
                 'Frame:': "NA", 
                 'Shaft:' : 'NA', 
                 'Weight / Grip Size:' : "NA", 
                 'Stringing Advice:' : "NA",
                 'Colour:' : "NA"}
        
        # je replie mon dico, les NA sont écrasés par les vraies données si elles existent 
        for des in range(len(tout_key)) :
            dico1[tout_key[des].text] = tout_value[des].text
       
        # Je mets mes données dans mon df en appelant les valeurs de mon dico
        liste_raquettes = liste_raquettes.append({"Name" : myName,
                                "Flex" : dico1['Flex:'],
                                "Frame" : dico1['Frame:'],
                                "Shaft" : dico1['Shaft:'],
                                "Weight / Grip Size" : dico1['Weight / Grip Size:'],
                                "Stringing Advice" : dico1['Stringing Advice:'],
                                "Color": dico1['Colour:']},
                               ignore_index=True)

        generate_number_delay()
        browser.back()
        generate_number_delay()
        
    return liste_raquettes
```

```python
# création de mon dataframe
liste_raq = pd.DataFrame({"Name" : [], 
                          "Flex" : [], 
                          "Frame" : [], 
                          "Shaft" : [], 
                          "Weight / Grip Size" : [],
                          "Stringing Advice" : [],
                          "Color": []
                         }
                        )

# je crée ma liste de type de raquette pour en connaitre la taille
family_raq = browser.find_elements_by_css_selector('ul.categoryCatalogue li')

# cette boucle passe de model en model 
for value in range(len(family_raq)) :
    family_raq = browser.find_elements_by_css_selector('ul.categoryCatalogue li')

    family_raq[value].click()
    
    # j'utilise ma fonction pour pour scraper le model de raquette 
    liste_raq = scap_page(liste_raq)
    
    # je teste s'il y a une deuximène page
    try : 
        myNext = browser.find_element_by_class_name('next')
        myNext.click()
        liste_raq = scap_page(liste_raq)
        browser.back()
    except : 
        print('pas de page 2')

    generate_number_delay()
    browser.back()
    generate_number_delay()
    
```

```python
#export du fichier en csv
liste_raq.to_csv("raquettes_yonex.csv", encoding='utf-8', index=False)
print("ok")
```
